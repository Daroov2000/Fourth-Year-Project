```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
from numpy import *
import numpy as np
from scipy.interpolate import interp1d
```

Base CDM using https://arxiv.org/pdf/1807.06209.pdf

```python
data0 = np.loadtxt("C:\\Users\\dhruv\\Desktop\\Uni Stuff\\MSci Project\\Data\\CDM.txt")

x0 = data0[:, 0]
y0 = data0[:, 1]

P0 = interp1d(x0,y0,'cubic',fill_value='extrapolate')
plot(x0,P0(x0),"b")

xscale('log')
yscale('log')

grid(color='black', linestyle='-', linewidth=1)
xlabel("k / [hMpc^-1]")
ylabel("P(k) / [Mpch^-1]^3")
title("CDM MPS")

show()
```

WDM with varying m_nCDM

```python
data1 = np.loadtxt("C:\\Users\\dhruv\\Desktop\\Uni Stuff\\MSci Project\\Data\\1000,0.7.txt")
data2 = np.loadtxt("C:\\Users\\dhruv\\Desktop\\Uni Stuff\\MSci Project\\Data\\3000,0.7.txt")
data3 = np.loadtxt("C:\\Users\\dhruv\\Desktop\\Uni Stuff\\MSci Project\\Data\\5000,0.7.txt")
data4 = np.loadtxt("C:\\Users\\dhruv\\Desktop\\Uni Stuff\\MSci Project\\Data\\7000,0.7.txt")
data5 = np.loadtxt("C:\\Users\\dhruv\\Desktop\\Uni Stuff\\MSci Project\\Data\\9000,0.7.txt")

x1 = data1[:, 0]
y1 = data1[:, 1]
x2 = data2[:, 0]
y2 = data2[:, 1]
x3 = data3[:, 0]
y3 = data3[:, 1]
x4 = data4[:, 0]
y4 = data4[:, 1]
x5 = data5[:, 0]
y5 = data5[:, 1]

P1 = interp1d(x1,y1,'cubic',fill_value='extrapolate')
P2 = interp1d(x2,y2,'cubic',fill_value='extrapolate')
P3 = interp1d(x3,y3,'cubic',fill_value='extrapolate')
P4 = interp1d(x4,y4,'cubic',fill_value='extrapolate')
P5 = interp1d(x5,y5,'cubic',fill_value='extrapolate')

plot(x1, P1(x1),'r', label = 'Mass=1 keV')
plot(x2, P2(x2),'g', label = 'Mass=3 keV')
plot(x3, P3(x3),'b', label = 'Mass=5 keV')
plot(x4, P4(x4),'y', label = 'Mass=7 keV')
plot(x5, P5(x5),'c', label = 'Mass=9 keV')

xscale('log')
yscale('log')

grid(color='black', linestyle='-', linewidth=1)
xlabel("k / [hMpc^-1]")
ylabel("P(k) / [Mpch^-1]^3")
title("WDM MPS for varying masses")

legend()
show()
```
